```python
# manually modified to fix any problems the Verification steps uncover
weekly = pd.read_csv("weeklySchedule{}.csv".format(SEASON), index_col=0).fillna("")
weekly.columns = weekly.columns.astype("int")
weekly
```

# Verify Integrity

```python
# check that every game has its pair entered correctly
for i, r in weekly.iterrows():
    print(i + ":")
    for c in range(1, 15+1):
        opp = weekly.at[i, c]
        print(opp)
        if opp != "":
            if "@" in opp:
                assert weekly.at[opp[1:], c] == i
            else:
                assert weekly.at[opp, c] == "@" + i
    print("---")
    print()
```

# Verify Total Counts

```python
weekly["Count"] = 15
for i, r in weekly.iterrows():
    weekly.at[i, "Count"] = 15 - r.value_counts()[""]
weekly.loc[weekly.Count != 9]
```

# Verify H/A Counts

```python
weekly["A_Count"] = 0
for i, r in weekly.iterrows():
    for opp in r.values:
        if type(opp) == str and "@" in opp:
            weekly.at[i, "A_Count"] += 1
    
weekly.loc[(weekly.A_Count < 4) & (weekly.A_Count != 0)].drop(columns=["Count"])
```

```python
# set IRL weeks
WEEK1 = datetime(2020, 7, 11, hour=12)
#est = pytz.timezone("America/New_York")

#dates = [None] + [est.localize(WEEK1 + timedelta(days=7*i)) for i in range(15)]
dates = [None] + [(WEEK1 + timedelta(days=7*i)) for i in range(15)]

dates
```

```python
games = set()

for w in range(1, 15+1):
    week = weekly[w]
    
    for t, o in week.items():
        if o:
            if "@" in o:
                games.add((t, o[1:], w))
            else:
                games.add((o, t, w))
```

```python
games
```

```python
for game in games:
    awaySchool = models.School.objects.get(abbreviation=game[0])
    homeSchool = models.School.objects.get(abbreviation=game[1])
    
    away = models.Team.objects.get(school=awaySchool, season=SEASON)
    home = models.Team.objects.get(school=homeSchool, season=SEASON)
    
    g = models.Game(away=away, home=home, datetime=dates[game[2]], season=SEASON, week=game[2])
    print(g)
    g.save()
```
