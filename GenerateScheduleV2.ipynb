```python
import os
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "cfbd.settings")
import django
django.setup()

from collections import defaultdict
import random

import pandas as pd

from main import models
```

```python
schools = models.School.objects.order_by("abbreviation")
#master = defaultdict(dict)
#master = pd.DataFrame(columns=[s.abbreviation for s in schools])

ncSlots = {}

rows = []
for school in schools:
    rows.append([0 for s in schools])
    ncSlots[school.abbreviation] = 3

master = pd.DataFrame(rows, columns=[s.abbreviation for s in schools])
master.index = [s.abbreviation for s in schools]
master
```

## Pre-Arranged NonCon Games

```python
# set pre-arranged noncon games here
prearranged = [
    ("UGA", "GT"),
    ("BAY", "MISS")
]

for matchup in prearranged:
    master[matchup[0]][matchup[1]] = 1
    master[matchup[1]][matchup[0]] = 1
    
    ncSlots[matchup[0]] -= 1
    ncSlots[matchup[1]] -= 1

master
```

## Randomly Fill Remaining NonCon Slots

```python
available = schools.all()

# todo: ensure nobody already has 3 before starting

for n in range(3):
    for school in schools:
        print(school)
        if ncSlots[school.abbreviation] >= 3-n:
            opponent = random.choice(available.exclude(conference=school.conference))
            print(opponent.abbreviation)
            
            master[school.abbreviation][opponent.abbreviation] = 1
            master[opponent.abbreviation][school.abbreviation] = 1
            
            ncSlots[school.abbreviation] -= 1
            if ncSlots[school.abbreviation] == 0:
                available = available.exclude(abbreviation=school.abbreviation)
                print("{} has filled all NC slots".format(school.abbreviation))
            ncSlots[opponent.abbreviation] -= 1
            if ncSlots[opponent.abbreviation] == 0:
                available = available.exclude(abbreviation=opponent.abbreviation)
                print("{} has filled all NC slots".format(opponent.abbreviation))
        
        else:
            print("Already has {} NC game(s)!".format(n+1))
            
        print()
    print()
    #break

master
```

```python
def OutputOpponentList(schools, n):
    for ab in [s.abbreviation for s in schools]:
        tot = 0
        print(ab)
        for k, v in master[ab].items():
            tot += v
            if v == 1:
                print("\t" + k)
        assert tot == n

OutputOpponentList(schools, 3)
```

## Divisional Opponents

```python
conferences = models.Conference.objects.all()

for conf in conferences:
    confSchools = schools.filter(conference__abbreviation=conf.abbreviation)
    print(conf)
    for s in confSchools:
        print("{} - {} {}".format(s, conf, s.division))
        divSchools = confSchools.filter(division=s.division)
        for o in divSchools:
            if o != s:
                master[s.abbreviation][o.abbreviation] = 1
                master[o.abbreviation][s.abbreviation] = 1
    
    print()
    
```
