```python
import os
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "cfbd.settings")
import django
django.setup()

from collections import defaultdict
import random

import pandas as pd
import numpy as np

from main import models
```

## Set up master matrix

```python
schools = models.School.objects.order_by("abbreviation")

rows = []
for school in schools:
    rows.append([0 for s in schools])

master = pd.DataFrame(rows, columns=[s.abbreviation for s in schools])
master.index = [s.abbreviation for s in schools]
master
```

```python
def OutputOpponentLists(schools, n=-1):
    for ab in [s.abbreviation for s in schools]:
        tot = 0
        print(ab)
        for k, v in master[ab].items():
            tot += v
            if v == 1:
                print("\t" + k)
        print(tot)
        
        if n > 0:
            assert tot == n
            
        print()
```

## Divisional Opponents

```python
conferences = models.Conference.objects.all()

for conf in conferences:
    if conf.abbreviation == "IND":
        continue
    
    confSchools = schools.filter(conference__abbreviation=conf.abbreviation)
    print(conf)
    for s in confSchools:
        print("{} - {} {}".format(s, conf, s.division))
        divSchools = confSchools.filter(division=s.division)
        for o in divSchools:
            if o != s:
                master[s.abbreviation][o.abbreviation] = 1
                master[o.abbreviation][s.abbreviation] = 1
    
    print()
```

```python
OutputOpponentLists(schools)
```

## Protected Cross-Division Rivals

```python
protected = [
    # SEC
    ("BAMA", "TENN"),
    ("AUB", "UGA"),
    ("ARK", "MIZZ"),
    ("LSU", "FLA"),
    ("MISS", "VAND"),
    ("MSST", "KENT"),
    ("TAMU", "SCAR"),
    
    # ACC
    ("BC", "VT"),
    ("CLEM", "GT"),
    ("FSU", "MIA"),
    ("LOU", "UVA"),
    ("NCST", "UNC"),
    ("CUSE", "PITT"),
    ("WAKE", "DUKE"),
    
    # B1G
    ("IND", "PURD"),
    
    # PAC
    ("UCLA", "CAL"),
    ("UCLA", "STAN"),
    ("USC", "CAL"),
    ("USC", "STAN"),
]
```

```python
for matchup in protected:
    master.at[matchup[0], matchup[1]] = 1
    master.at[matchup[1], matchup[0]] = 1
```

## Remaining Cross-Division Rivals

```python
def ScheduleCrossDivision(cAbb):
    
    confSchools = schools.filter(conference__abbreviation=cAbb)
    print(cAbb)
    
    # create list of available games
    remList = []
    for s in confSchools:
        n = 9 - int(master[s.abbreviation].sum())
        for i in range(n):
            remList.append(s.abbreviation)
    print(remList)
    
    gameList = []
    
    while len(remList) > 0:    
        # take first team
        t1 = np.random.choice(remList)
        
        # filter list for only valid opponents (different school, opposite division, not in schedule already)
        valList = [t for t in remList if t != t1 and 
                                         schools.get(abbreviation=t).division != schools.get(abbreviation=t1).division and
                                         (t1, t) not in gameList and (t1, t) not in protected and (t, t1) not in protected]
        #print(t1, valList)
        
        # pick random opponent, save in gameList
        t2 = np.random.choice(valList)
        gameList.append((t1, t2))
        gameList.append((t2, t1))
        
        # remove from remList
        remList.remove(t1)
        remList.remove(t2)
        
        print(t1, t2)
        print(len(remList))
    
    # save game in Master
    for g in gameList:
        master.at[g[0], g[1]] = 1
    
    print()
```

```python
ScheduleCrossDivision("ACC")
```
